```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python

# Function to read .bin image
def read_bin_image(filename, shape=(256, 256)):
    with open(filename, 'rb') as f:
        img = np.frombuffer(f.read(), dtype=np.uint8)
    return img.reshape(shape)
```

```python
# Problem 1: Read and display Lena and Peppers images
lena = read_bin_image('lena.bin')
peppers = read_bin_image('peppers.bin')

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(lena, cmap='gray')
plt.title("Lena Image")

plt.subplot(1,2,2)
plt.imshow(peppers, cmap='gray')
plt.title("Peppers Image")
plt.show()
```

```python
# Create image J
J = np.hstack((lena[:, :128], peppers[:, 128:]))
plt.imshow(J, cmap='gray')
plt.title("Image J (Half Lena, Half Peppers)")
plt.show()

# Create image K by swapping halves
K = np.hstack((J[:, 128:], J[:, :128]))
plt.imshow(K, cmap='gray')
plt.title("Image K (Swapped Halves)")
plt.show()
```

```python
# Problem 2: Read lenagray.jpg and create its negative
J1 = cv2.imread('lenagray.jpg', cv2.IMREAD_GRAYSCALE)
J2 = 255 - J1  # Negative image

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(J1, cmap='gray')
plt.title("Original Lenagray")

plt.subplot(1,2,2)
plt.imshow(J2, cmap='gray')
plt.title("Negative Lenagray")
plt.show()

cv2.imwrite('negative_lenagray.jpg', J2)
```

```python
# Problem 3: Read lena512color.jpg and swap color channels
J1 = cv2.imread('lena512color.jpg')
J2 = J1.copy()
J2[:, :, 0] = J1[:, :, 2]  # Swap Red with Blue
J2[:, :, 1] = J1[:, :, 0]  # Swap Green with Red
J2[:, :, 2] = J1[:, :, 1]  # Swap Blue with Green

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(J1, cv2.COLOR_BGR2RGB))
plt.title("Original Lena Color")

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(J2, cv2.COLOR_BGR2RGB))
plt.title("Color Swapped Lena")
plt.show()

cv2.imwrite('swapped_lena.jpg', J2)
```
