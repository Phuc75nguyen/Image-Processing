Obtain the image “johnny.bin” from the course web site.  This  image  has  256      256 
pixels. Each pixel has 8 bits. Plot the histogram of the original image. Write a program to perform histogram equalization on this image.  Show the equalized image and plot its histogram.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#hàm đọc ảnh johny.bin
def read_bin_img(filename, shape = (256,256)):
    with open(filename, 'rb') as f:
        img = np.frombuffer(f.read(), dtype=np.uint8)
    return img.reshape(shape)


#đọc ảnh gốc
johnny = read_bin_img("johnny.bin")
```

```python
#vẽ histogram của ảnh gốc
plt.Figure(figsize=(12,6))

plt.subplot(2,2,1)
plt.imshow(johnny, cmap='gray')
plt.title("Ảnh gốc")
```

```python
plt.subplot(2, 2, 2)
plt.hist(johnny.ravel(), bins=256, range=[0,256], color='black')
plt.title("Histogram of Original Image")
```

```python
#Cân bằng lược đồ Histogram Equalization
johnny_equalized = cv2.equalizeHist(johnny)
```

```python
#Hiển thị ảnh sau khi cân bằng và vẽ histogram mới
plt.subplot(2, 2, 3)
plt.imshow(johnny_equalized, cmap='gray')
plt.title("Equalized Image")

plt.subplot(2, 2, 4)
plt.hist(johnny_equalized.ravel(), bins=256, range=[0,256], color='black')
plt.title("Histogram of Equalized Image")

plt.tight_layout()
plt.show()
```
