Full-Scale Contrast Stretching

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
#đọc ảnh bin
def read_bin_image(filename, shape=(256, 256)): 
    with open(filename, 'rb')    as f:
        img = np.frombuffer(f.read(), dtype=np.uint8)
    return img.reshape(shape)


def contrast_stretching(image):
    min_val = np.min(image)
    max_val = np.max(image)
    stretched = ((image - min_val) / (max_val - min_val) * 255).astype(np.uint8)
    return stretched

lady = read_bin_image("lady.bin")
stretched_lady = contrast_stretching(lady)

plt.imshow(stretched_lady, cmap='gray')
plt.title("Contrast Stretched Image")
plt.show()
```
