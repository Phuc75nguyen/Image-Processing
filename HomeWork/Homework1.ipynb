```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ƒê·ªçc ·∫£nh grayscale
lena = cv2.imread('lenagray.jpg', cv2.IMREAD_GRAYSCALE)
parot = cv2.imread('parrot.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Ki·ªÉm tra xem ·∫£nh c√≥ load th√†nh c√¥ng kh√¥ng
if lena is None or parot is None:
    raise FileNotFoundError("Kh√¥ng t√¨m th·∫•y file ·∫£nh. H√£y ki·ªÉm tra l·∫°i ƒë∆∞·ªùng d·∫´n!")
```

```python
# üîπ 1. Histogram Equalization (AHE - Adaptive Histogram Equalization)
lena_hist_eq = cv2.equalizeHist(lena)
parot_hist_eq = cv2.equalizeHist(parot)
```

```python
# üîπ 2. Contrast-Limited Adaptive Histogram Equalization (CLAHE)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
lena_clahe = clahe.apply(lena)
parot_clahe = clahe.apply(parot)
```

```python
# Hi·ªÉn th·ªã k·∫øt qu·∫£
fig, axes = plt.subplots(2, 3, figsize=(12, 8))

# Lena Images
axes[0, 0].imshow(lena, cmap='gray')
axes[0, 0].set_title('Original Lena')

axes[0, 1].imshow(lena_hist_eq, cmap='gray')
axes[0, 1].set_title('Histogram Equalization (Lena)')

axes[0, 2].imshow(lena_clahe, cmap='gray')
axes[0, 2].set_title('CLAHE (Lena)')

# Parrot Images
axes[1, 0].imshow(parot, cmap='gray')
axes[1, 0].set_title('Original Parrot')

axes[1, 1].imshow(parot_hist_eq, cmap='gray')
axes[1, 1].set_title('Histogram Equalization (Parrot)')

axes[1, 2].imshow(parot_clahe, cmap='gray')
axes[1, 2].set_title('CLAHE (Parrot)')

# ·∫®n tr·ª•c
for ax in axes.flatten():
    ax.axis('off')

plt.tight_layout()
plt.show()
```
