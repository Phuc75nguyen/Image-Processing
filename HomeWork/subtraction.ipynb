```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
live = cv2.imread('live.jpg')
mask = cv2.imread('mask.jpg')
```

```python
subtraction = cv2.subtract(live, mask)
```

```python
# Hin th kt qu vi Matplotlib
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(live, cmap="gray")
plt.title("Live Image")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(mask, cmap="gray")
plt.title("Mask Image")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(subtraction, cmap="gray")
plt.title("Subtracted Result")
plt.axis("off")

plt.tight_layout()
plt.show()
```
