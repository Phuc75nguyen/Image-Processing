The  images  in  problems  1-5  are  all  of  size  8 × 8

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Hàm in ma trận 8x8 theo định dạng ascii với 4 chữ số sau dấu phẩy
def print_ascii_array(arr, title):
    print(title)
    # Làm tròn giá trị của mảng đầu vào đến 4 chữ số sau dấu phẩy
    rounded = np.round(arr, 4)
    for row in rounded:
        # định dạng mỗi phần tử của hàng thành chuỗi có 8 ký tự, trong đó 4 ký tự sau dấu phẩy
        print(" ".join(f"{num:8.4f}" for num in row))
    print("\n")
# tạo lưới chỉ số với hệ thống zero-based: các hàng và cột từ 0 đến 7
cols, rows = np.meshgrid(np.arange(8), np.arange(8))
# Hằng số dùng trong công thức tính 2pi/8
two_pi_over_8 = 2 * np.pi / 8
```

```python
#----------------Problem 1 ------------------
# công thức I1 = 0.5 * exp(j*2π/8*(2*COLS + 2*ROWS))
I1 = 0.5 * np.exp(1j * two_pi_over_8 * (2*cols + 2*rows))
I1_real = I1.real
I1_imag = I1.imag

# tính toán DFT của I1 và dịch chuyển tần số 0 về trung tâm
I1_dft = np.fft.fftshift(np.fft.fft2(I1))
I1_dft_real = I1_dft.real
I1_dft_imag = I1_dft.imag

print_ascii_array(I1_real, "Problem 1: Real Part of I1 (8×8)")
print_ascii_array(I1_imag, "Problem 1: Imaginary Part of I1 (8×8)")
print_ascii_array(I1_dft_real, "Problem 1: Centered DFT of I1 - Real Part (8×8)")
print_ascii_array(I1_dft_imag, "Problem 1: Centered DFT of I1 - Imaginary Part (8×8)")

# hiển thị I1 và phần thực, ảo của DFT của I1
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(I1_real, cmap='gray')
plt.title("I1 - Real Part")
plt.colorbar()
plt.subplot(1,2,2)
plt.imshow(I1_imag, cmap='gray')
plt.title("I1 - Imaginary Part")
plt.colorbar()
plt.suptitle("Problem 1: I1")
plt.show()
```

```python
# ------------------ Problem 2 ------------------
# công thức I2 = 0.5 * exp(-j*2π/8*(2*COLS + 2*ROWS))
I2 = 0.5 * np.exp(-1j * two_pi_over_8 * (2*cols + 2*rows))
I2_real = I2.real
I2_imag = I2.imag

# tính toán DFT của I2 và dịch chuyển tần số 0 về trung tâm
I2_dft = np.fft.fftshift(np.fft.fft2(I2))
I2_dft_real = I2_dft.real
I2_dft_imag = I2_dft.imag

print_ascii_array(I2_real, "Problem 2: Real Part of I2 (8×8)")
print_ascii_array(I2_imag, "Problem 2: Imaginary Part of I2 (8×8)")
print_ascii_array(I2_dft_real, "Problem 2: Centered DFT of I2 - Real Part (8×8)")
print_ascii_array(I2_dft_imag, "Problem 2: Centered DFT of I2 - Imaginary Part (8×8)")
```

```python
# ------------------ Problem 3 ------------------
# công thức I3 = cos(2π/8*(2*COLS + 2*ROWS))
# lưu ý: Since I1 = 0.5*exp(jθ) and I2 = 0.5*exp(-jθ), I1+I2 = cos(θ)
I3 = np.cos(two_pi_over_8 * (2*cols + 2*rows))
I3_real = I3  # I3 is real-valued
I3_imag = np.zeros_like(I3)

# tính toán DFT của I3 và dịch chuyển tần số 0 về trung tâm
# DFT của một tín hiệu thực là một tín hiệu chẵn, do đó phần ảo của DFT của I3 là 0
I3_dft = np.fft.fftshift(np.fft.fft2(I3))
I3_dft_real = I3_dft.real
I3_dft_imag = I3_dft.imag

print_ascii_array(I3_real, "Problem 3: I3 (Real Image, 8×8)")
print_ascii_array(I3_dft_real, "Problem 3: Centered DFT of I3 - Real Part (8×8)")
print_ascii_array(I3_dft_imag, "Problem 3: Centered DFT of I3 - Imaginary Part (8×8)")
```

```python
# ------------------ Problem 4 ------------------
# công thức I4 = sin(2π/8*(2*COLS + 2*ROWS))
I4 = np.sin(two_pi_over_8 * (2*cols + 2*rows))
I4_real = I4  # I4 is real-valued
I4_imag = np.zeros_like(I4)

# tính toán DFT của I4 và dịch chuyển tần số 0 về trung tâm
# DFT của một tín hiệu thực là một tín hiệu chẵn, do đó phần ảo của DFT của I4 là 0
I4_dft = np.fft.fftshift(np.fft.fft2(I4))
I4_dft_real = I4_dft.real
I4_dft_imag = I4_dft.imag

print_ascii_array(I4_real, "Problem 4: I4 (Real Image, 8×8)")
print_ascii_array(I4_dft_real, "Problem 4: Centered DFT of I4 - Real Part (8×8)")
print_ascii_array(I4_dft_imag, "Problem 4: Centered DFT of I4 - Imaginary Part (8×8)")
```

```python
# ------------------ Problem 5 ------------------
# ở bài 5, tần só thay đổi u1 = 1.5, v1 = 1.5 cpi
# công thứ I5 = cos(2π/8*(1.5*COLS + 1.5*ROWS))
I5 = np.cos(two_pi_over_8 * (1.5*cols + 1.5*rows))
I5_real = I5  # chỉ có phần thực
I5_imag = np.zeros_like(I5)

# tính toán DFT của I5 và dịch chuyển tần số 0 về trung tâm
# DFT của một tín hiệu thực là một tín hiệu chẵn, do đó phần ảo của DFT của I5 là 0
I5_dft = np.fft.fftshift(np.fft.fft2(I5))
I5_dft_real = I5_dft.real
I5_dft_imag = I5_dft.imag
# in ma trận I5 và các thành phần DFT của nó
print_ascii_array(I5_real, "Problem 5: I5 (Real Image, 8×8)")
print_ascii_array(I5_dft_real, "Problem 5: Centered DFT of I5 - Real Part (8×8)")
print_ascii_array(I5_dft_imag, "Problem 5: Centered DFT of I5 - Imaginary Part (8×8)")

# hiển thị I5 và phần thực của DFT của I5
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(I5, cmap='gray', vmin=-1, vmax=1)
plt.title("I5 - Real Image")
plt.colorbar()
plt.subplot(1,2,2)
plt.imshow(I5_dft_real, cmap='gray')
plt.title("I5 DFT - Real Part")
plt.colorbar()
plt.suptitle("Problem 5")
plt.show()
```
